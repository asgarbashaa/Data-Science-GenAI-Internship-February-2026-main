```python
def smart_parking_lot(capacity, logs):
    parked = 0
    peak = 0
    
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
        
        if parked > peak:
            peak = parked
    
    print("Currently Parked Vehicles:", parked)
    
    if parked > capacity:
        print("Parking Status: Full - Alert!")
    else:
        print("Parking Status: Available")
    
    print("Peak Usage:", peak)


smart_parking_lot(50, ["IN","IN","IN","OUT","IN","IN","OUT"])
```

```python
def delivery_time_estimator(distance, traffic, weather):
    time = distance * 5  
    
    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 8
    
    if weather == "Rainy":
        time += 10
    elif weather == "Storm":
        time += 20
    
    print("Estimated Delivery Time:", time, "minutes")


delivery_time_estimator(8, "High", "Rainy")
```

```python
def theatre_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100
    
    print("Occupancy:", round(occupancy), "%")
    
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


theatre_occupancy(200, [1]*150)
```

```python
def server_load_classifier(cpu_readings):
    avg_cpu = sum(cpu_readings) / len(cpu_readings)
    
    print("Average CPU Load:", round(avg_cpu), "%")
    
    if avg_cpu < 50:
        print("Server Status: Normal")
    elif 50 <= avg_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


server_load_classifier([45,60,70,85,90])
```

```python
def classroom_resource_monitor(resources):
    overused = []
    
    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)
    
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


classroom_resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
def event_registration(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


event_registration(100, 105)
```
